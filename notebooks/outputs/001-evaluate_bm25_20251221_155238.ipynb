# RAG Retrieval Evaluation: Dense vs Hybrid Search

Notebook để đánh giá so sánh hiệu quả của Dense-only search vs Hybrid search (Dense + BM25 Sparse) sử dụng ranx metrics.

```python
# Parameters (injected by papermill if running via script)
TEST_DATA_PATH = "../data/test_queries_evaluation.json"
K = 10
RUN_TIMESTAMP = None
```

## 1. Setup Retriever và Comparator

```python
retriever = ContextRetriever()
evaluator = RetrievalEvaluator(retriever=retriever)
```

## 2. Example: Single Query Evaluation

Test với 1 query để hiểu output format

```python
result = evaluator.compare(
    query="Tỉnh Điện Biên có chính sách gì để thu hút nguồn nhân lực trình độ cao?",
    relevant_doc_ids=["537/QĐ-UBND"],
    k=10
)
```

```python
print("DENSE-ONLY SEARCH")
for metric, value in result["dense"].items():
    if metric != "retrieved_docs":
        print(f"{metric:20s}: {value:.4f}")
print(f"\nTop 5 docs: {result['dense']['retrieved_docs']}")
```

```python
print("HYBRID SEARCH (Dense + Sparse)")
for metric, value in result["hybrid"].items():
    if metric != "retrieved_docs":
        print(f"{metric:20s}: {value:.4f}")
print(f"\nTop 5 docs: {result['hybrid']['retrieved_docs']}")
```

```python
print("IMPROVEMENT (%)")
for metric, value in result["improvement"].items():
    print(f"{metric:25s}: {value:+.2f}%")

print(f"\nWinner: {result['winner'].upper()}")
```

## 3. Batch Evaluation: Multiple Queries

Evaluate trên nhiều queries để có kết quả aggregate

```python
# Load test queries from JSON file (using parameter from papermill)
import json

with open(TEST_DATA_PATH, "r", encoding="utf-8") as f:
    test_queries = json.load(f)

# Display loaded queries
print(f"Loaded {len(test_queries)} test queries from: {TEST_DATA_PATH}")
for i, q in enumerate(test_queries[:3], 1):
    print(f"\n{i}. Query: {q['query'][:80]}...")
    print(f"   Relevant docs: {q['relevant_doc_ids']}")
    print(f"   Category: {q.get('category', 'N/A')}")
```

```python
# Run batch evaluation with parameter K
summary = evaluator.compare_batch(test_queries, k=K)
```

```python
print("=" * 60)
print(f"EVALUATION SUMMARY ({summary['num_queries']} queries)")
print("=" * 60)

print("\nAverage Metrics:")
print("\nDense-only:")
for metric, value in summary["dense_avg"].items():
    print(f"  {metric:20s}: {value:.4f}")

print("\nHybrid:")
for metric, value in summary["hybrid_avg"].items():
    print(f"  {metric:20s}: {value:.4f}")

print("\nAverage Improvement:")
for metric, value in summary["avg_improvement"].items():
    print(f"  {metric:25s}: {value:+.2f}%")

print("\nWins:")
print(f"  Hybrid: {summary['wins']['hybrid']} ({summary['wins']['win_rate_hybrid']:.1f}%)")
print(f"  Dense:  {summary['wins']['dense']} ({100 - summary['wins']['win_rate_hybrid']:.1f}%)")

print(f"\nRecommendation: {summary['recommendation'].upper()}")
```

```python
## 4. Prepare Data for Visualization
```

```python
metrics_data = []

for metric_name, dense_value in summary["dense_avg"].items():
    hybrid_value = summary["hybrid_avg"][metric_name]
    metrics_data.append({"Metric": metric_name, "Method": "Dense", "Score": dense_value})
    metrics_data.append({"Metric": metric_name, "Method": "Hybrid", "Score": hybrid_value})

df_metrics = pd.DataFrame(metrics_data)
df_metrics
```

## 5. Bar Chart Comparison

```python
fig = px.bar(
    df_metrics,
    x="Metric",
    y="Score",
    color="Method",
    barmode="group",
    title="Dense vs Hybrid Search: Average Metrics",
    labels={"Score": "Average Score"},
    color_discrete_map={"Dense": "#636EFA", "Hybrid": "#EF553B"}
)
fig.update_layout(height=500)
fig.show()
```

## 6. Radar Chart: Overall Comparison

```python
# Radar chart for overall comparison
metrics = list(summary["dense_avg"].keys())
dense_values = [summary["dense_avg"][m] for m in metrics]
hybrid_values = [summary["hybrid_avg"][m] for m in metrics]

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
    r=dense_values,
    theta=metrics,
    fill='toself',
    name='Dense-only',
    line_color='blue'
))

fig.add_trace(go.Scatterpolar(
    r=hybrid_values,
    theta=metrics,
    fill='toself',
    name='Hybrid',
    line_color='red'
))

fig.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 1]
        )
    ),
    showlegend=True,
    title="Dense vs Hybrid: Radar Comparison"
)

fig.show()
```

## 7. Per-Query Analysis

```python
# Analyze per-query results
per_query_data = []

for i, result in enumerate(summary["details"], 1):
    per_query_data.append({
        "Query": f"Q{i}",
        "Query Text": result["query"][:50] + "...",
        "Winner": result["winner"],
        "Dense NDCG@10": result["dense"]["ndcg@10"],
        "Hybrid NDCG@10": result["hybrid"]["ndcg@10"],
        "Improvement %": result["improvement"]["ndcg@10_pct"]
    })

df_per_query = pd.DataFrame(per_query_data)
df_per_query
```

## 8. NDCG Improvement Visualization

```python
fig = px.bar(
    df_per_query,
    x="Query",
    y="Improvement %",
    title="NDCG@10 Improvement per Query (Hybrid vs Dense)",
    labels={"Improvement %": "NDCG@10 Improvement (%)"},
    text="Improvement %",
    color="Improvement %",
    color_continuous_scale="RdYlGn"
)
fig.update_traces(texttemplate='%{text:.1f}%', textposition='outside')
fig.update_layout(height=500)
fig.show()
```

## 9. So sánh 3 phương pháp: Dense vs Hybrid TFIDF vs Hybrid BM25

<span id="papermill-error-cell" style="color:red; font-family:Helvetica Neue, Helvetica, Arial, sans-serif; font-size:2em;">Execution using papermill encountered an exception here and stopped:</span>

```python
from llm_engineering.domain.queries import EmbeddedQuery
from llm_engineering.application.networks.sparse_embedding import SparseEmbeddingModel
from llm_engineering.application.networks.sparse_encoder.tfidf import TFIDFSparseEncoder
from llm_engineering.application.networks.sparse_encoder.mb25 import BM25SparseEncoder

def evaluate_all_methods(query: str, relevant_doc_ids: list[str], k: int = 10):
    query_id = f"q_{hash(query) % 10000}"
    results = {}

    # Dense
    EmbeddedQuery.Config.use_sparse_vector_index = False
    dense_docs = [r.document_number for r in retriever.search(query, k=k, expand_to_n_queries=1) if r.document_number]
    results['dense'] = evaluator.evaluate_query(query_id, dense_docs, relevant_doc_ids, k)

    # Hybrid TFIDF
    EmbeddedQuery.Config.use_sparse_vector_index = True
    SparseEmbeddingModel._instance = None
    SparseEmbeddingModel.algorithm = TFIDFSparseEncoder
    tfidf_docs = [r.document_number for r in retriever.search(query, k=k, expand_to_n_queries=1) if r.document_number]
    results['hybrid_tfidf'] = evaluator.evaluate_query(query_id, tfidf_docs, relevant_doc_ids, k)

    # Hybrid BM25
    SparseEmbeddingModel._instance = None
    SparseEmbeddingModel.algorithm = BM25SparseEncoder
    bm25_docs = [r.document_number for r in retriever.search(query, k=k, expand_to_n_queries=1) if r.document_number]
    results['hybrid_bm25'] = evaluator.evaluate_query(query_id, bm25_docs, relevant_doc_ids, k)

    return results
```

```python
all_results = [evaluate_all_methods(q['query'], q['relevant_doc_ids'], K) for q in test_queries]

methods = ['dense', 'hybrid_tfidf', 'hybrid_bm25']
avg_metrics = {}

for method in methods:
    metric_names = [k for k in all_results[0][method].keys()]
    avg_metrics[method] = {m: sum(r[method][m] for r in all_results) / len(all_results) for m in metric_names}

print(f"Evaluated {len(all_results)} queries\n")
for method in methods:
    print(f"{method.upper().replace('_', ' ')}:")
    for metric, value in avg_metrics[method].items():
        print(f"  {metric:15s}: {value:.4f}")
    print()
```

```python
comparison_data = []
for method in methods:
    for metric, value in avg_metrics[method].items():
        comparison_data.append({"Metric": metric, "Method": method.replace('_', ' ').title(), "Score": value})

df_comp = pd.DataFrame(comparison_data)

fig = px.bar(df_comp, x="Metric", y="Score", color="Method", barmode="group",
             title="So sánh Dense vs Hybrid TFIDF vs Hybrid BM25",
             color_discrete_map={"Dense": "#636EFA", "Hybrid Tfidf": "#EF553B", "Hybrid Bm25": "#00CC96"},
             height=500)
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
metric_names = list(avg_metrics['dense'].keys())

fig = go.Figure()
for method, color in [('dense', 'blue'), ('hybrid_tfidf', 'red'), ('hybrid_bm25', 'green')]:
    fig.add_trace(go.Scatterpolar(
        r=[avg_metrics[method][m] for m in metric_names],
        theta=metric_names,
        fill='toself',
        name=method.replace('_', ' ').title(),
        line_color=color
    ))

fig.update_layout(
    polar=dict(radialaxis=dict(visible=True, range=[0, 1])),
    title="Radar Chart: 3-Way Comparison"
)
fig.show()
```

```python
ndcg_data = []
for i, r in enumerate(all_results, 1):
    ndcg_data.append({
        "Query": f"Q{i}",
        "Dense": r['dense']['ndcg@10'],
        "Hybrid TFIDF": r['hybrid_tfidf']['ndcg@10'],
        "Hybrid BM25": r['hybrid_bm25']['ndcg@10']
    })

df_ndcg = pd.DataFrame(ndcg_data)

fig = go.Figure()
for method, color in [('Dense', 'blue'), ('Hybrid TFIDF', 'red'), ('Hybrid BM25', 'green')]:
    fig.add_trace(go.Scatter(
        x=df_ndcg['Query'], y=df_ndcg[method],
        mode='lines+markers', name=method,
        line=dict(color=color, width=2), marker=dict(size=8)
    ))

fig.update_layout(title='NDCG@10: Dense vs Hybrid TFIDF vs Hybrid BM25',
                  xaxis_title='Query', yaxis_title='NDCG@10', height=500, hovermode='x unified')
fig.show()
```
